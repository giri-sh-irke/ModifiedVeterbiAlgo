### Build the vanilla Viterbi based POS tagger

```python
# t X v matrix
t = len(set(tags))
v = len(vocab)
print("Tag length : ",t)
print("Vocab length : ",v)
t_v_mat = np.zeros((t, v))
```

```python
# Train test spliting. 5% will be considered as test_set or validation_set
train_set, test_set = train_test_split(nltk_data,test_size=0.05)
train_tag_words = [pair for sent in train_set for pair in sent]
test_tag_words = [pair for sent in test_set for pair in sent]

print(len(train_tag_words))
print(len(test_tag_words))
```

```python
# Emission Probability
def word_given_tag(word, tag, train_bag = train_tag_words):
    tag_list = [pair for pair in train_bag if pair[1]==tag]
    count_tag = len(tag_list)
    w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word]
    count_w_given_tag = len(w_given_tag_list)
    
    return (count_w_given_tag, count_tag)
```

```python
# Transition Probability
def t2_given_t1(t2, t1, train_bag = train_tag_words):
    tags = [pair[1] for pair in train_bag]
    count_t1 = len([t for t in tags if t==t1])
    count_t2_t1 = 0
    for index in range(len(tags)-1):
        if tags[index]==t1 and tags[index+1] == t2:
            count_t2_t1 += 1
    return (count_t2_t1, count_t1)
```

```python
# NOUN given ADJECTIVE was the previous tag
t2_given_t1('NOUN','ADJ')
```

```python
# NOUN given DETERMINANT was the previous tag
t2_given_t1('NOUN','DET')
```

```python
len(train_tag_words)
```

```python
# Constructing a tags_matrix (tXt)
T = set(tags)
tags_matrix = np.zeros((len(T), len(T)), dtype='float32')
for i, t1 in enumerate(list(T)):
    for j, t2 in enumerate(list(T)): 
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1]
```

```python
# Constructing a dataFrame from the tags_matrix
tags_df = pd.DataFrame(tags_matrix, columns = list(T), index=list(T))
```

#### Vanilla Viterbi function

```python
# Vanilla Viterbi Heuristic
def Vanilla_Viterbi(words, train_bag = train_tag_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        state_max = T[p.index(pmax)] 
        state.append(state_max)
    return list(zip(words, state))
```

```python
random.seed(1234)

# choose random 5 sents
rndom = [random.randint(1,len(test_set)) for x in range(5)]

# list of sents
test_run = [test_set[i] for i in rndom]

# list of tagged words
test_run_base = [tup for sent in test_run for tup in sent]

# list of untagged words
test_tagged_words = [tup[0] for sent in test_run for tup in sent]
```

```python
start = time.time()
tagged_seq = Vanilla_Viterbi(test_tagged_words)
end = time.time()
difference = end-start
```

### Solve the problem of unknown words

#### First modification : For the unknown words, the emission probablity comes out to be zero. Hence considering only the Transition probabilies for the same

```python
# Modified Viterbi Algorithm - 1 
def Mod_Viterbi_1(words, train_bag = train_tag_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        p_transition = []
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            p_transition.append(transition_p)
            
        pmax = max(p)
        if(pmax == 0):
            pmax = max(p_transition)
            state_max = T[p_transition.index(pmax)]
        else:    
        # getting state for which probability is maximum
            state_max = T[p.index(pmax)] 
        state.append(state_max)
    return list(zip(words, state))
```

```python
start = time.time()
tagged_seq = Mod_Viterbi_1(test_tagged_words)
end = time.time()
difference = end-start
print("Time elapsed : ", difference)
check = [i for i, j in zip(tagged_seq, test_run_base) if i == j] 
accuracy = len(check)/len(tagged_seq)
print('accuracy', accuracy*100)
```

**Observation :**
Increase in the accuracy as the first tag is not assigned to the unknown words 

```python
# Mapping the incorrect tags with their correct tags
[j for i, j in enumerate(zip(tagged_seq, test_run_base)) if j[0] != j[1]]
```

#### Second Modification - Rule based tagging

```python
patterns = [
    (r'.*ing$', 'VERB'),              # gerund
    (r'.*ed$', 'VERB'),               # past tense
    (r'.*es$', 'VERB'),               # 3rd singular present
    (r'.*\'s$', 'NOUN'),              # possessive nouns
    (r'.*s$', 'NOUN'),                # plural nouns
    (r'^-?[0-9]+(.[0-9]+)?$', 'NUM'), # cardinal numbers
    (r'\*T?\*?-?[0-9]*$', 'X'),       # Regex for the unknown tags
    (r'^[A-Z][a-z].*', 'NOUN'),       # Tagging capital lettered names
    (r'.*', 'NOUN')                    # nouns
]
```

```python
regexp_tagger = nltk.RegexpTagger(patterns)
```

```python
# Modified Viterbi Heuristic
def Mod_Viterbi_2(words, train_bag = train_tag_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        if(pmax == 0):
            unknown_tagged_words = regexp_tagger.tag([word])[0][1]
            state.append(unknown_tagged_words)
        else:
            state_max = T[p.index(pmax)] 
            state.append(state_max)
    return list(zip(words, state))
```

```python
start = time.time()
tagged_seq = Mod_Viterbi_2(test_tagged_words)
end = time.time()
difference = end-start
```
